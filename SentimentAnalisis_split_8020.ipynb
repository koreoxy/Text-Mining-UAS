```python
data = dataset.rename({"Instagram Comment Text":"Comment"}, axis="columns")
data.head()
```

# **TAHAPAN PREPROCESSING**
tahapan Preprocessing bertujuan untuk membersihkan data text.

```python
# Tahapan 1
# proses case folding
# case folding digunakan untuk membuat huruf menjadi kecil semua atau lowercasae
import re
def casefolding(Comment):
  Comment = Comment.lower()
  Comment = Comment.strip(" ")
  Comment = re.sub(r'[?|$|.|!²_:")(-+,]', '', Comment)
  return Comment
data['Comment'] = data['Comment'].apply(casefolding)
data.head(100)
```

```python
#Tahapan 2
#Proses Tokenizing
#Tokenizing digunakan untuk membuat setiap kata menjadi berurutan dari 1 kalimat
#atau memisahkan setiap kata-kata dengan tanda koma di dalam satu kalimat
def token(Comments):
  nstr = Comments.split(' ')
  dat= []
  a = -1
  for hu in nstr:
    a = a + 1
  if hu == '':
    dat.append(a)
  p = 0
  b = 0
  for q in dat:
    b = q - p
    del nstr[b]
    p = p + 1
  return nstr
data['Comment'] = data['Comment'].apply(token)
data.head(10)
```

```python
#Tahap 3
#Proses Filtering
#Stopwords_removal menghilangkan kata-kata yang tidak bermanfaat atau kata-kata hubung contoh "yang" "ke"
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

def stopword_removal(Comments):
  filtering = stopwords.words('indonesian', 'english')
  x = []
  data = []
  def myFunc(x):
    if x in filtering:
      return False
    else:
      return True
  fit = filter(myFunc, Comments)
  for x in fit:
    data.append(x)
  return data
data['Comment'] = data['Comment'].apply(stopword_removal)
data.head()
```

```python
#Tahap 4
#Proses Stemming
#Stemming dalam bahasa inggris digunakan untuk mengubah suatu kata menjadi kata dasarnya 
#Stemming dalam bahasa indonesia digunakan untuk menghilangkan kata himbuan misalkan mencuci menjadi cuci
!pip install Sastrawi
from sklearn.pipeline import Pipeline
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(Comments):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for w in Comments:
    dt = stemmer.stem(w)
    do.append(dt)
  d_clean = []
  d_clean = " ".join(do)
  print(d_clean)
  return d_clean
data['Comment'] = data['Comment'].apply(stemming)

data.to_csv('data_clean.csv', index=False)
data_clean = pd.read_csv('data_clean.csv', encoding='latin1')
data_clean.head()
```

akurasi suatu algoritma atau mesin itu tergantung dari preprocessing datanya, kalau semakin bersih data tersebut di preprocessing maka data tersebut semakin bagus tingkat akurasi nya.

```python
data_clean = data_clean.astype({'Sentiment' : 'category'})
data_clean = data_clean.astype({'Comment': 'string'})
data_clean.dtypes
```

```python
#Proses TF-IDF
#TF-IDF ini digunakan untuk membobotti suatu kata, semakin banyak kata didalam suatu dokumen maka semakin tinggi nilai bobotnya
#misalkan didalam suatu dokumen memiliki kata yang berulang contohnya kata uang muncul berulang kali maka kata uang tersebut memiliki bobot yang tinggi dibandingkan kata yang lain.
#Proses TF-IDF ini adalah proses default yang digunakan dalam text mining karena dibutuhkan untuk pembobotan kata

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
text_tf = tf.fit_transform(data_clean['Comment'].astype('U'))
text_tf
```

preprocessing text telah selesai sampai proses TF-IDF

```python
#Splitting data
#Spliting data digunakan untuk memisahkan antara data training dan data testing, data testing dan data training dibagi menjadi 20, 80
#Data testing nya 0,2 persen menjadi 20%
#dan Data training nya menjadi 80%
#dan random state 42 adalah urutan acak atau random pembagian data testing dan data training. 
#supaaya data testing dan training memiliki jenis atau kategory sentiment yang berbeda-beda dan hasil data training tersebut akan menjadi bagus
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text_tf, data_clean['Sentiment'], test_size = 0.2, random_state = 42)
```

```python
#Performa algoritma Naive Bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

clf = MultinomialNB().fit(X_train, y_train)
predicted = clf.predict(X_test)
print("MultinomialNB Accuracy:", accuracy_score(y_test,predicted))
print("MultinomialNB Precision:", precision_score(y_test, predicted, average="binary", pos_label="negative"))
print("MultinomialNB Recall:", recall_score(y_test, predicted, average="binary", pos_label="negative"))
print("MultinomialNB f1_score:", f1_score(y_test, predicted, average="binary", pos_label="negative"))

print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('======================================================\n')
print(classification_report(y_test, predicted, zero_division=0))
```
